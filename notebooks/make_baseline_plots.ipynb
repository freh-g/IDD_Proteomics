# Calculates network features 

```python
def calculate_network_metrics(G):
    # Check if the graph is empty
    if len(G) == 0:
        return "The graph is empty."
    
    # Average Degree
    degrees = [degree for node, degree in G.degree()]
    average_degree = np.mean(degrees)
    
    # Clustering Coefficient
    clustering_coefficient = nx.average_clustering(G)
    
    # Network Density
    density = nx.density(G)
    
    # Node Connectivity
    connectivity = nx.node_connectivity(G)
    
    # Modularity
    partition = cl.best_partition(G)
    modu = cl.modularity(partition, G)

    # Output the results
    metrics = {
        "Average Degree": average_degree,
        "Clustering Coefficient": clustering_coefficient,
        "Network Density": density,
        "Modularity": modu
    }
    
    return metrics


def minmax_scale_list(values, new_min=0, new_max=1):
    # Check if the list is empty
    if not values:
        return []
    
    # Compute the minimum and maximum values in the list
    old_min = min(values)
    old_max = max(values)
    
    # Handle edge case where all values are the same
    if old_min == old_max:
        # If all values are the same, scaling does not change values
        return [new_min + (new_max - new_min) / 2 for _ in values]
    
    # Apply Min-Max scaling formula
    scaled_values = [
        new_min + (value - old_min) / (old_max - old_min) * (new_max - new_min)
        for value in values
    ]
    
    return scaled_values




def minmax_scale_multiple_dicts(dicts):
    scaled_dicts = []
    keys = dicts[0].keys()
    dict_of_values = dict(zip(keys,[[] for k in keys]))
    # collect all the data from the initial list of dict
    for d in dicts:
        for ke in d.keys():
            # append the specific key of the dict in the initial list of dicts to the temporary dictionary
            dict_of_values[ke].append(d[ke])
    
    # scale the values

    for k,v in dict_of_values.items():
        dict_of_values[k] = minmax_scale_list(v)

    # Distribute the values across initial dictionaries

    for i,d in enumerate(dicts):
        for k,v in d.items():
            d[k] = dict_of_values[k][i]
    
    return(dicts)


    

    
```

```python
networks = []
path = '../outputs/baseline_networks/pickle_nets/'
nets = os.listdir(path)
for n in nets:
    with open(path+n,'rb') as f:
        network = pi.load(f)
    networks.append(network)

# Calculate metrics for each network
metrics_list = [calculate_network_metrics(G) for G in networks]
# metrics_list = minmax_scale_multiple_dicts(metrics_list)
# Metric names and corresponding values for each network
metric_names = ["Average Degree", "Clustering Coefficient", "Network Density","Modularity"]
metric_values = [list(metrics.values()) for metrics in metrics_list]

# Plotting
x = np.arange(len(metric_names))  # Label locations
width = 0.15  # Width of the bars

fig, ax = plt.subplots(figsize=(12, 8))

for i, values in enumerate(metric_values):
    ax.bar(x + i * width, values, width, label=nets[i].replace('.pickle',''))

# Add labels, title, and legend
ax.set_xlabel('Metrics')
ax.set_ylabel('Values')
ax.set_title('Comparison of Network Metrics Across 5 Networks')
ax.set_xticks(x + width * 2)
ax.set_xticklabels(metric_names)
ax.legend()
plt.savefig("../outputs/baseline_networks/metrics.jpeg",dpi = 300)

plt.show()
```
