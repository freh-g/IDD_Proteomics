```python
import graphlot as gp
import pandas as pd
import networkx as nx
import os
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable, get_cmap
```

```python
interactome = pd.read_csv('../inputs/interactome.sif',names=['source','score','target'])
```

```python
interactome_net = nx.from_pandas_edgelist(interactome[['source','target']])
```

```python
gene_symbol_list_seeds=['IL6', 'IL6ST','IL16', 'IL1B', 'IL17D', 'IL18', 'IL20', 'TNF', 'LIF', 'OSM', 'CCL2', 'CCL3', 'CCL5', 'CCL7', 'CXCL1', 'CXCL2', 'CXCL3', 'CXCL8', 'CX3CL1', 'IGF1', 'GDF5', 'TGFB1', 'GDF6', 'PMP2', 'VEGFA']
path='../outputs/secretome/'
for file in os.listdir('../outputs/secretome/'):
    if file != 'plots':
        with open(path+file,'r') as f:
            o = f.readlines()
    pctg = 0.1
    figurename = file.replace('.txt','.jpg')
    onepercent = int(0.1*len(o))
    newlist = o[-onepercent:]
    listofgenes = [a.split('\t')[0] for a in newlist]
    subnet = nx.subgraph(interactome_net,listofgenes)
    # add nodes attributes
    seedict = dict(zip(listofgenes,['seed' if n in gene_symbol_list_seeds else 'non_seed' for n in listofgenes]))
    nx.set_node_attributes(subnet,seedict,'seed')
    gp.visualize_network(subnet,node_color_attribute='seed',legend='seed',node_annotation=True,figure_size=(20,15),save = f'../outputs/secretome/plots/{figurename}')
   

    
    
```
