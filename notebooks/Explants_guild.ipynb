```python
import pandas as pd
import graphlot as grpl
import genopyc as gp
import networkx as nx
import matplotlib_venn as pltv
import os
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable, get_cmap
```

```python
def minmax_scale_list(data, min_value=0, max_value=1):
    """
    Min-max scale a list of numbers to a specified range.

    Parameters:
        data (list): The list of numbers to be scaled.
        min_value (float): The minimum value of the desired range (default: 0).
        max_value (float): The maximum value of the desired range (default: 1).

    Returns:
        list: The scaled list of numbers.
    """
    min_data = min(data)
    max_data = max(data)
    scaled_data = [(x - min_data) / (max_data - min_data) * (max_value - min_value) + min_value for x in data]
    return scaled_data
```

# Load and clean the data

```python
explants = pd.read_excel('../data/Explants data.xlsx')
explants = explants.iloc[22:].set_index('treatment').drop('Analyte',axis = 1)
explants['RETN'] = explants['RETN ']
explants.drop('RETN ',inplace=True,axis=1)
explants_proteins = explants.columns.tolist()
interactome = pd.read_csv('../data/HIPPIE-current.mitab.txt',sep='\t')
interactome_proteins = set(interactome['Gene Name Interactor A'].tolist() + interactome['Gene Name Interactor B'].tolist())
interactome_proteins = [p for p in list(interactome_proteins) if type(p) != float]
```

# Create the interactome enriched file 

```python
# Create the network from hippie keeping the lcc

interactome_net = nx.from_pandas_edgelist(interactome[['Gene Name Interactor A','Gene Name Interactor B']],source='Gene Name Interactor A',target='Gene Name Interactor B')
largest_cc = max(nx.connected_components(interactome_net), key=len)
interactome_cc = nx.subgraph(interactome_net,largest_cc)

# Detect the shortes paths between all nodes of the explants

shortest_paths = {}
for node1 in explants_proteins:
    for node2 in explants_proteins:
        if node1 != node2:
            shortest_paths[(node1, node2)] = nx.shortest_path(interactome_net, source=node1, target=node2)

# Detect all the nodes in the shortest paths

sp_nodes = set()
for path in shortest_paths.values():
    sp_nodes.update(path)
sp_nodes = [n for n in sp_nodes if type(n) != float]

# Make the subnetwork
subnet_enriched = nx.subgraph(interactome_net,sp_nodes)

# Save the enriched subnetwork

subnet_enriched_sif = nx.to_pandas_edgelist(subnet_enriched)
subnet_enriched_sif.insert(1,'score',1)
subnet_enriched_sif.to_csv('../inputs/explants/interactome.sif',index=None,header=None,sep = ' ')

```

# Normalize and save nodes in files

```python
path = '../inputs/explants/'
experiments = ['control',"IL4","IL10","IL1B","IL1\+IL4"]

for i,exp in enumerate(experiments):
    df = explants[explants.index.str.match(exp)]
    display(df.head())
    secretome_proteins = df.columns.tolist()
    means = [df[col].mean() for col in df.columns.tolist()]
    scaled_means = minmax_scale_list(means)
    
    # Make a dictionary of cytokynes and values
    dictionary_of_means = dict(zip(df.columns.tolist(),scaled_means))
    
    # Force manually depending on the experiment
    if i == 1:
        dictionary_of_means['IL4'] = 1
    elif i == 2:
        dictionary_of_means['IL10'] = 1
    elif i == 3:
        dictionary_of_means['IL1B'] = 1
    elif i == 4:
        dictionary_of_means['IL1'] = 1
        dictionary_of_means['IL4'] = 1
        exp = "IL1IL4"


    for node in sp_nodes:
        if node not in secretome_proteins:
            dictionary_of_means[node] = 0
    
    # Create the folder
    if not os.path.isdir(path):
        os.makedirs(path)
    # Write the files
    with open(path + 'explants_' +exp+'_nodes.txt','w') as f:
        for gene,score in dictionary_of_means.items():
            f.write(gene.rstrip() + ' ' + str(score) + ' ' + '\n') 
```
